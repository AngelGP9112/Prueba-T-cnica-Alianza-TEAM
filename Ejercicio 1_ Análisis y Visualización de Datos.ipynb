```python
# Contar el número total de filas duplicadas
num_duplicados = data.duplicated().sum()
print(f"Número de filas duplicadas encontradas: {num_duplicados}")
```

```python
# Ver las filas duplicadas (esto mostrará todas las filas que tienen duplicados, incluyendo la primera aparición usando keep=False)
data[data.duplicated(keep=False)]
```

```python
# Eliminar duplicados (por defecto, en todas las columnas y manteniendo la primera aparición)
data.drop_duplicates(inplace=True)
```

```python
# Validación Columnas Numéricas
print("--- 1. Estado inicial de las columnas numéricas con decimales ---")
print(data[['precio_unitario', 'costo_unitario', 'margen_unitario']].head())
print(data[['precio_unitario', 'costo_unitario', 'margen_unitario']].dtypes)

# Lista de columnas que necesitan ser convertidas a numérico (float)
columnas_a_numerico = ['precio_unitario', 'costo_unitario', 'margen_unitario']

for col in columnas_a_numerico:
    # Solo procesa si la columna no es ya numérica
    if not pd.api.types.is_numeric_dtype(data[col]):
        print(f"Convirtiendo columna: {col}")
        # Eliminar el separador de miles (puntos)
        # Reemplazar la coma decimal por un punto decimal
        # Convertir a numérico (float). 'errors="coerce"' convertirá cualquier valor problemático a NaN.
        data[col] = data[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
        data[col] = pd.to_numeric(data[col], errors='coerce')
    else:
        print(f"La columna '{col}' ya es numérica. No se requiere conversión.")

print("\n--- 2. Estado de las columnas después de la conversión a numérico ---")
print(data[['precio_unitario', 'costo_unitario', 'margen_unitario']].head())
print(data[['precio_unitario', 'costo_unitario', 'margen_unitario']].dtypes)

# Verificar si se generaron NaNs durante la conversión (valores que no se pudieron convertir)
for col in columnas_a_numerico:
    if data[col].isnull().sum() > 0:
        print(f"ADVERTENCIA: Se encontraron {data[col].isnull().sum()} valores nulos en '{col}' después de la conversión.")
        # Considerar cómo manejar estos NaNs: rellenar con la media/mediana o eliminar las filas.
        # Por ejemplo, para rellenar con la mediana:
        # data[col].fillna(data[col].median(), inplace=True)
```

```python
# Redondear cifra Margen unitario
# La columna 'margen_unitario' ya es de tipo numérico (float), segun comprobación

print("\n--- Redondeando y reemplazando 'margen_unitario' a 2 decimales ---")

# Reemplazar la columna original 'margen_unitario' con su versión redondeada a 2 decimales
data['margen_unitario'] = data['margen_unitario'].round(2)

print("\nPrimeras 5 filas de la columna 'margen_unitario' después del redondeo y reemplazo:")
print(data['margen_unitario'].head())
```

```python
# impresion de datos en dataset
print(data.shape)
data.head()
```

```python
# Limpieza de columnas categoricas: categoria, tipo_producto_producto

columnas_categoricas = ['categoria', 'tipo_producto', 'producto']

print("--- 2. Limpieza de Columnas Categóricas ---")

for col in columnas_categoricas:
    print(f"\nProcesando columna: {col}")
    print(f"Valores únicos antes de limpiar {col}:")
    print(data[col].unique())

    # a) Eliminar espacios en blanco al inicio y al final
    data[col] = data[col].astype(str).str.strip()

    # b) Unificar a un mismo caso (ej. Título, para que 'producto A' sea 'Producto A')
    data[col] = data[col].str.title() # Convierte la primera letra de cada palabra a mayúscula

    # c) Corregir errores de codificación o tipográficos conocidos
    if col == 'tipo_producto':
        # Reemplazar 'Pago Â´nico' por 'Pago Único'
        data[col] = data[col].str.replace('Pago Â´Nico', 'Pago Único', regex=False)
        data[col] = data[col].str.replace('Pâ°go Â´Nico', 'Pago Único', regex=False) # Por si hay otra variante

    print(f"Valores únicos después de limpiar {col}:")
    print(data[col].unique())

    # Converción a tipo 'category' para optimizar memoria y velocidad de procesamiento
    data[col] = data[col].astype('category')
    print(f"Tipo de dato de {col} después de la limpieza y conversión a category: {data[col].dtype}")
```

```python
# impresion de datos en dataset
print(data.shape)
data.head()
```

```python
# Gerarquizacón columna fecha, calculos posteriores

print("--- INSPECCIÓN INICIAL DE LA COLUMNA 'fecha' ---")
# Muestra el tipo de dato actual de la columna 'fecha'
print(f"Tipo de dato actual de 'fecha': {data['fecha'].dtype}")
```

```python
# Muestra algunas fechas crudas tal como están cargadas (primeras 10, para ver si hay inconsistencias)
print("\nPrimeros 10 valores de la columna 'fecha' (sin procesar):")
print(data['fecha'].head(10))
```

```python
# Muestra el conteo de valores únicos en la columna 'fecha' para detectar formatos extraños o valores no fecha
print("\nConteo de valores únicos en 'fecha' (solo los primeros 20):")
print(data['fecha'].value_counts().head(20))
```

```python
# impresion de datos en dataset
print(data.shape)
data.head()
```

```python
# 1. Se crea una columna TEMPORAL de tipo datetime a partir de la columna 'fecha' original.
#    Uso de 'errors="coerce"' para convertir valores no válidos a NaT.
#    El 'format' es CRÍTICO: debe coincidir con el formato REAL de tus fechas en el CSV.
#    Según el formato del archivo csv es 'YYYY-MM-DD'.
data['fecha_datetime_temp'] = pd.to_datetime(data['fecha'], format='%Y-%m-%d', errors='coerce')

# 2. Reverción de la columna 'fecha' ORIGINAL  a tipo string con el formato deseado.
#    Esto garantiza que data['fecha'] SIEMPRE se vea como 'YYYY-MM-DD' en el DataFrame.
data['fecha'] = data['fecha_datetime_temp'].dt.strftime('%Y-%m-%d')

# Imprimir información para verificar el estado de la columna 'fecha'
print("--- Paso 1: Columna 'fecha' restaurada a string YYYY-MM-DD ---")
print(f"Tipo de dato de la columna 'fecha': {data['fecha'].dtype}")
print("Primeras 5 filas de la columna 'fecha':")
print(data['fecha'].head())

# La columna 'fecha_datetime_temp' ahora contiene la versión datetime y será usada en los siguientes pasos.
```

```python
# impresion de datos en dataset
print(data.shape)
data.head()
```

```python
# Creación de la columna 'año' extrayendo el año de la columna temporal datetime.
if 'fecha_datetime_temp' in data.columns and data['fecha_datetime_temp'].dtype == 'datetime64[ns]':
    data['año'] = data['fecha_datetime_temp'].dt.year
    print("--- Paso 2: Columna 'año' (numérica) creada ---")
    print(f"Tipo de dato de la columna 'año': {data['año'].dtype}")
    print("\nPrimeras 5 valores de la columna 'año':")
    print(data['año'].head())
else:
    print("Error: 'fecha_datetime_temp' no existe o no es de tipo datetime. Ejecuta el Paso 1 primero.")
```

```python
# Creación mes numérico extraído de 'fecha_datetime_temp'.

if 'fecha_datetime_temp' in data.columns and data['fecha_datetime_temp'].dtype == 'datetime64[ns]':
    data['mes'] = data['fecha_datetime_temp'].dt.month
    print("--- Paso 3.5: Columna 'mes' (numérica) creada ---")
    print(f"Tipo de dato de la columna 'mes': {data['mes'].dtype}")
    print("\nPrimeros 5 valores de la columna 'mes':")
    print(data['mes'].head())
else:
    print("Error: 'fecha_datetime_temp' no existe o no es de tipo datetime. Ejecuta el Paso 1 primero.")
```

```python
# Columna mes_nombre, para el nombre del mes
if 'fecha_datetime_temp' in data.columns and data['fecha_datetime_temp'].dtype == 'datetime64[ns]':
    # Diccionario para mapear números de mes a nombres en español
    nombres_meses = {
        1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril',
        5: 'Mayo', 6: 'Junio', 7: 'Julio', 8: 'Agosto',
        9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
    }

    # Extraer el número del mes de la columna temporal y mapearlo al nombre.
    data['mes_nombre'] = data['fecha_datetime_temp'].dt.month.map(nombres_meses)

    print("--- Paso 3: Columna 'mes_nombre' (como palabra en español) creada ---")
    print(f"Tipo de dato de la columna 'mes_nombre': {data['mes_nombre'].dtype}")
    print("\nPrimeras 5 valores de la columna 'mes_nombre':")
    print(data['mes_nombre'].head())
else:
    print("Error: 'fecha_datetime_temp' no existe o no es de tipo datetime. Ejecuta el Paso 1 primero.")
```

```python
# Creación de la columna 'dia' extrayendo el día del mes de la columna temporal datetime.
if 'fecha_datetime_temp' in data.columns and data['fecha_datetime_temp'].dtype == 'datetime64[ns]':
    data['dia'] = data['fecha_datetime_temp'].dt.day
    print("--- Paso 5: Columna 'dia' (numérica del mes) creada ---")
    print(f"Tipo de dato de la columna 'dia': {data['dia'].dtype}")
    print("\nPrimeras 5 valores de la columna 'dia':")
    print(data['dia'].head())
else:
    print("Error: 'fecha_datetime_temp' no existe o no es de tipo datetime. Ejecuta el Paso 1 primero.")
```

```python
# Creación de la columna 'dia_nombre' con el nombre correspondiente al día de la semana.
if 'fecha_datetime_temp' in data.columns and data['fecha_datetime_temp'].dtype == 'datetime64[ns]':
    # Diccionario para mapear números de día de la semana (0=Lunes, 6=Domingo) a nombres en español
    nombres_dias_semana = {
        0: 'Lunes', 1: 'Martes', 2: 'Miércoles', 3: 'Jueves',
        4: 'Viernes', 5: 'Sábado', 6: 'Domingo'
    }

    # Extraer el número del día de la semana de la columna temporal y mapearlo al nombre.
    data['dia_nombre'] = data['fecha_datetime_temp'].dt.weekday.map(nombres_dias_semana)

    print("--- Paso 4: Columna 'dia_nombre' (como palabra en español) creada ---")
    print(f"Tipo de dato de la columna 'dia_nombre': {data['dia_nombre'].dtype}")
    print("\nPrimeras 5 valores de la columna 'dia_nombre':")
    print(data['dia_nombre'].head())
else:
    print("Error: 'fecha_datetime_temp' no existe o no es de tipo datetime. Ejecuta el Paso 1 primero.")
```

```python
# Eliminación de la columna temporal fecha_datetime_temp del datetime
if 'fecha_datetime_temp' in data.columns:
    data.drop(columns=['fecha_datetime_temp'], inplace=True)
    print("--- Paso 6: Columna 'fecha_datetime_temp' eliminada ---")
else:
    print("La columna 'fecha_datetime_temp' no existe para ser eliminada.")

# Para verificar el estado final de las columnas relacionadas con fecha
print("\n--- Estado final de las columnas de fecha en el DataFrame ---")
print(data[['fecha', 'año', 'mes_nombre', 'dia_nombre', 'dia']].head())
```
